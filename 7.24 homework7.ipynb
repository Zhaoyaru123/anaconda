```python
def getNumber(s):
    N1 = ""
    for i in s :
        if i.isalpha():
            if "abcABC".find(i) != -1: 
                N1 += "2"
            if "defDEF".find(i) != -1: 
                N1 += "3"
            if "ghiGHI".find(i) != -1: 
                 N1 += "4"
            if "jklJKL".find(i) != -1: 
                 N1 += "5"
            if "mnoMNO".find(i) != -1: 
                 N1 += "6"
            if "pqrsPQRS".find(i) != -1: 
                 N1 += "7"
            if "tuvTUV".find(i) != -1: 
                 N1 += "8"
            if "wxyzWXYZ".find(i) != -1: 
                 N1 += "9"
        else:
            N1 += i
    return N1
```
